## 失業率xml——>json

- age

```python
import subprocess
import json
import os

# 運行 un_age_process.py
script_path = os.path.join("un_age_process.py")
subprocess.run(["python", script_path], check=True)

# 讀取生成的 JSON 文件
json_path = os.path.join("unemployment", "taiwan_unemployment_age.json")
with open(json_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# 打印基礎信息
print("=" * 50)
print("資料基礎信息")
print("=" * 50)
print(f"總條目數量: {len(data)}")
print(f"資料欄位: {list(data[0].keys()) if data else '無資料'}")
print(f"欄位數量: {len(data[0].keys()) if data else 0}")

# 打印前3筆資料
print("\n" + "=" * 50)
print("前3筆資料")
print("=" * 50)
for i, record in enumerate(data[:3], 1):
    print(f"\n第 {i} 筆:")
    print(json.dumps(record, ensure_ascii=False, indent=2))
```
